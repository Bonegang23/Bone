# 🎬 Bone Productions Logo Animation
This notebook uses MoviePy to generate a logo animation video from static frames.

```python
from moviepy.editor import *

# Load static frames (make sure these are uploaded to this directory)
frame_files = ['frame1.png', 'frame2.png', 'frame3.png']  # Add more if needed
clips = [ImageClip(f).set_duration(1) for f in frame_files]  # 1 second each

# Concatenate frames and add optional audio (if available)
video = concatenate_videoclips(clips, method='compose')
video.write_videofile("logo_animation.mp4", fps=24)
```
